```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df = df[~df['train_model'].str.startswith('Bus', na=False)]
```

```python
df['hbf'] = df['hbf'].fillna('Unknown')
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df['scheduled_time'] = pd.to_datetime(df['date'].astype(str) + " " + df['scheduled_time'].astype(str), errors='coerce')
df['expected_time'] = pd.to_datetime(df['date'].astype(str) + " " + df['expected_time'].astype(str), errors='coerce')
```

```python
df['scheduled_time'] = df['scheduled_time'].dt.tz_localize(None)
df['expected_time'] = df['expected_time'].dt.tz_localize(None)
```

```python
df['scheduled_time'] = df['scheduled_time'].dt.strftime('%Y-%m-%d %H:%M')
df['expected_time'] = df['expected_time'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df['real_time_cleaned'] = df['real_time_due_to_delay'].astype(str).str.extract(r'^(\d{2}:\d{2})')
print(df[['real_time_due_to_delay', 'real_time_cleaned']])
```

```python
df.head()
```

```python
df.tail()
```

```python
# Define the exact column order you want
df = df[['date', 'hbf', 'scheduled_time', 'expected_time', 'train_model', 'route',
         'platform', 'real_time_due_to_delay', 'real_time_cleaned', 'has_delay']]
```

```python
df.tail()
```

```python
df['scheduled_time'] = pd.to_datetime(df['scheduled_time'], errors='coerce')
df['expected_time'] = pd.to_datetime(df['expected_time'], errors='coerce')

df['delay_minutes'] = (df['expected_time'] - df['scheduled_time']).dt.total_seconds() / 60

```

```python
df.head()
```

```python
df.tail()
```

```python
# Finding maximum and minimum value for further analysis

df['delay_minutes'] = df['delay_minutes'].apply(lambda x: max(x, 0) if pd.notna(x) else x)

print("Minimum delay:", round(df['delay_minutes'].min(), 2), "minutes")
print("Maximum delay:", round(df['delay_minutes'].max(), 2), "minutes")
```

```python
# categorizing delay
df['Delay_Type'] = df['delay_minutes'].apply(
    lambda x: 'No Delay' if x == 0 else
              'Short Delay' if x <= 2 else
              'Moderate Delay' if x <= 4 else
              'Unknown'
)
```

```python

# extracting platform_number from platform as platform column contains string value alon with platform number.
import re

df['platform'] = df['platform'].astype(str).str.strip()
df['Platform_Number'] = df['platform'].str.extract(r'^(\d+)')
df['Platform_Number'] = df['Platform_Number'].replace('', np.nan)
df['Platform_Number'] = df['Platform_Number'].astype('Int64')
df[['platform', 'Platform_Number']].head(10)
```

Understanding Delays

```python
from pandasql import sqldf

q = """
SELECT has_delay, COUNT(*) AS count
FROM df
GROUP BY has_delay
"""
sqldf(q)
```

```python
# Checking the distribution of delays by minutes
df['has_delay'] = df['delay_minutes'] > 0
duckdb.sql("""
SELECT
    delay_minutes,
    COUNT(*) AS total_trains,
    SUM(CASE WHEN has_delay THEN 1 ELSE 0 END) AS total_delayed,
    ROUND(SUM(CASE WHEN has_delay THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS delay_rate
FROM df
GROUP BY delay_minutes
ORDER BY delay_minutes
""").df()
```

```python
duckdb.sql("""
SELECT hbf, SUM(has_delay) AS total_delays
FROM df
GROUP BY hbf
ORDER BY total_delays DESC
""").df()
```
